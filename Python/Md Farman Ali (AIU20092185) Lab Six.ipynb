```python
graph = {
'A' : ['B','C'],
'B' : ['D', 'E'],
'C' : ['F', 'G'],
'D' : [],
'E' : [],
'F' : [],
'G' : []
}
path = list() #LIST CONSTRUCTOR,EMTY LIST
def DFS(currentNode,destination,graph,maxDepth,curList):
    curList.append(currentNode)#adds the current node to the current [ath being traversed]
    if currentNode==destination:
        return True #reture;ff the current node is the path to the list
    if maxDepth<=0:
        path.append(curList)
        return False
    for node in graph[currentNode]:#for each node connected to the courrent node in the graph
        if DFS(node,destination,graph,maxDepth-1,curList):
            return True #find the destination node in the search
        else:
            curList.pop() #else remove the last element from the curlist
            #since it did not lead to a path to the destination node
    return False
def iterativeDDFS(currentNode,destination,graph,maxDepth):
    for i in range(maxDepth):
        curList = list() # each iteration, a new emty list is created
        if DFS(currentNode,destination,graph,i,curList):#recursive dfs function
            return True
    return False
#cheacks if there exists a path from node 'A'to node 'E'in group with the maxdeapth
if not iterativeDDFS('A','E',graph,1):
    print("Path is not available")
else:
    print("Path exists")
    print(path.pop())
if not iterativeDDFS('A','G',graph,3):
    print("Path is not available")
else:
    print("Path exists")
    print(path.pop())
```

```python
from collections import deque #deque ended queqe
#allows both ends to add or remove elements
class Node:
    def __init__(self, val, neighbors=[]):# initialise new node
        self.val = val
        self.neighbors = neighbors
        self.visited_right = False
        self.visited_left = False
        self.parent_right = None
        self.parent_left = None
def bidirectional_search(s, t): #define the bidirectional search
    def extract_path(node):
        node_copy = node
        path = [] #empty list
        while node:#start traversing
            path.append(node.val)
            node = node.parent_right
        path.reverse()# parent node
        del path[-1]
        while node_copy:
            path.append(node_copy.val)
            node_copy = node_copy.parent_left
        return path
    q = deque([])# assign q with double ended queque
    q.append(s)#start node
    q.append(t)#detination node
    s.visited_right = True
    t.visited_left = True
    while len(q) > 0:
        n = q.pop()
        if n.visited_left and n.visited_right:
            return extract_path(n)
        for node in n.neighbors:
            if n.visited_left == True and not node.visited_left:
                node.parent_left = n
                node.visited_left = True
                q.append(node)
            if n.visited_right == True and not node.visited_right:
                node.parent_right = n
                node.visited_right = True
                q.append(node)
    return False
n0 = Node('A')
n1 = Node('B')
n2 = Node('C')
n3 = Node('D')
n4 = Node('E')
n5 = Node('F')
n6 = Node('G')
n0.neighbors = []
n1.neighbors = [n0]
n2.neighbors = [n0]
n3.neighbors = [n1]
n4.neighbors = [n1]
n5.neighbors = [n2]
n6.neighbors = [n2]
print(bidirectional_search(n4, n6))
print(bidirectional_search(n3, n6))
print(bidirectional_search(n0, n6))
```
